```python
# Define the list of FITS file paths
from astropy.io import fits

fits_files = []

for i in range(40, 100, 1):
    fits_files.append(f'WASP-12b/WASP-12b_000{i}_bsdf.fits')

for i in range(100, 270, 1):
    fits_files.append(f'WASP-12b/WASP-12b_00{i}_bsdf.fits')
    
print(fits_files)
```

```python
# collect data from fits files
fits_data = []
    
for fits_file in fits_files:
    file_data = fits.open(fits_file)
    image_data = file_data[0].data
   
    fits_data.append(image_data)

print(fits_data[0:5])
```

```python
# define position of star and aperture
from photutils.aperture import CircularAperture

position = [1729.8084, 4077.6048]
aperture = CircularAperture(position,r=10.0)


# perform aperture photometry on selected star for each fits file. Append the aperture sums to a list
from photutils.aperture import aperture_photometry
star_data = []

for file in fits_data:
    star_data.append(aperture_photometry(file,aperture)['aperture_sum'])
    
print(star_data[0:5])
```

```python
#plot aperture sums
import matplotlib.pyplot as plt
plt.plot(star_data)
plt.ylabel('Aperture Sum')
plt.xlabel('Time')
plt.show()
```

```python
#plot aperture sums
import matplotlib.pyplot as plt
plt.plot(star_data, 'o')
plt.ylabel('Aperture Sum')
plt.xlabel('Time')
plt.show()
```
